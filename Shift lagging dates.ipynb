```python
import pandas as pd
df= pd.read_csv('stocks2.csv',parse_dates=['Date'], index_col='Date')
df
```

```python
df.shift(1) #shifts the prices by 1 day
```

```python
df.shift(-1)
```

```python
df['Prev Day Price'] = df['price'].shift(1)
df
```

```python
df['Price Change'] = df.price - df['Prev Day Price']
df
```

```python
df['5 day Return'] = (df['price']-df['price'].shift(5))*100/df['price'].shift(5)
df
```
